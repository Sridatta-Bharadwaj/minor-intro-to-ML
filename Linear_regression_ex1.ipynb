```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/chicago_taxi_trips_2016_01.csv')
```

```python
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.isnull().sum()
```

```python
data_cleaned = data[['trip_miles', 'trip_total']].dropna()
x = data_cleaned['trip_miles'].values.reshape(-1, 1)
y = data_cleaned['trip_total'].values
```

```python
data_cleaned.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, model.predict(x_train), color = 'red')
plt.title('trip distance vs trip cost (training set)')
plt.xlabel('trip distance in miles')
plt.ylabel('trip total in dollers')
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_train, model.predict(x_train), color = 'red')
plt.title('trip distance vs trip cost (testing set)')
plt.xlabel('trip distance in miles')
plt.ylabel('trip total in dollers')
plt.show()
```
